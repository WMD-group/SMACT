```python
from smact.screening import smact_filter, SmactFilterOutputs
from smact.io.elementembeddings import composition_featuriser, species_composition_featuriser
import smact
import pprint

# Define a list of elements to screen
elements = ["Ti", "Ga", "O"]

# Convert element symbols into smact.Element objects
element_objects = [smact.Element(e) for e in elements]

# Apply smact_filter to find valid compositions within a stoichiometry threshold
allowed_combinations = smact_filter(element_objects, threshold=6, return_output=SmactFilterOutputs.formula)

# Display the allowed combinations in a readable format
feature_vectors = composition_featuriser(
    allowed_combinations,
)

for formula, fv in zip(allowed_combinations, feature_vectors):
    print(f"Formula: {formula}, Vector: {fv} \n")
```
